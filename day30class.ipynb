```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
path = 'https://raw.githubusercontent.com/ValRCS/Python_SheGoesTech_22/main/data/california_housing.csv'
df=pd.read_csv(path)
df.head()
#The target variable is the median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).

```

```python
X = df[['HouseAge']]
y= df[['target']]
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
```

```python
X_train[:5], y_train[:5]
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
lr.score(X_test, y_test) #so if it's very low does it mean I need to change the feature?
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='magenta')
plt.show()
```

```python
lr.fit(X2_train, y2_train)
y2_pred = lr.predict(X2_test)
```
